```python
import json
with open("/content/students.json", "r") as file:
  data = json.load(file)
print("JSON file loaded")
ref = db.reference("students")
for student in data["students"]:
  ref.child(str(student["id"])).set(student)
print("Data uploaded successfully!")
```

```python
import requests
import json
import firebase_admin
from firebase_admin import credentials, db
# Step 1: Download the key from GitHub
url = "https://raw.githubusercontent.com/HerlieVinceBDiente/Quarter3-Codes/refs/heads/main/students.json"
response = requests.get(url)
# Step 2: Save it locally in Colab
with open("/content/firebase_key (1).json", "wb") as f:
  f.write(response.content)
# Step 3: Initialize Firebase
# Check if the default Firebase app is already initialized
if not firebase_admin._apps:
  cred = credentials.Certificate("/content/firebase_key (1).json")
  firebase_admin.initialize_app(cred, {
  'databaseURL': 'https://grade8-json-db-96a63-default-rtdb.asia-southeast1.firebasedatabase.app/'
  })
  print("Connected to Firebase!")
else:
  print("Firebase app already initialized.")
#make sure the url is the raw file from the github
```
